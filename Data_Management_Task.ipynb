Question:1  
 	
We have collected feedback from the session which was conducted online for Domain Helpers only. So, we will be considering the feedback of only the Domain Helpers.
There can be some duplicate entries in the feedback as well which we need to take care of as well.
Once the data is prepared and cleared we will have to create charts which can represent the data which is present on the sheet.

```python
domain_helpers_emails=domain_helpers["Email"].unique()
filtered_feedback=Feedback_data[Feedback_data["Email Address"].isin(domain_helpers_emails)]

#Removing the duplicates:
filtered_feedback = filtered_feedback.drop_duplicates(subset=["Email Address"], keep="last")

# Calculate average rating per activity
activity_columns = ["Rate the activity-1", "Rate the following activity-2", 
                    "Rate the following activity-3", "Rate the following activity-4"]

# Calculate averages for each activity and overall
average_ratings = filtered_feedback[activity_columns].mean()
overall_average = filtered_feedback["Rate your experience with the overall event"].mean()

print("Average Ratings per Activity:")
print(average_ratings)
print(f"Overall Average Rating: {overall_average}")


# Bar chart for average activity ratings
plt.figure(figsize=(10, 6))
average_ratings.plot(kind="bar", color="skyblue")
plt.title("Average Ratings per Activity for Domain Helpers")
plt.xlabel("Activities")
plt.ylabel("Average Rating")
plt.show()
```

# Overall Average Rating:4.19



The chart illustrates that all activities received relatively high ratings, with Activity 3 rated the highest on average at 4.47. This indicates positive feedback overall, suggesting that participants had a good experience in each of the activities.
